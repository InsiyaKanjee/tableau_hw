since Jan data-set is different, the lookback window used is Feb 2021- Aug 2021, next appending all to create one large DF

```python
Feb_data = Feb_data.append(Mar_data, ignore_index=True)
```

```python
Feb_data = Feb_data.append(Apr_data, ignore_index=True)
```

```python
Feb_data = Feb_data.append(Jun_data, ignore_index=True)
```

```python
Feb_data = Feb_data.append(Jul_data, ignore_index=True)
```

```python
Feb_data = Feb_data.append(Aug_data, ignore_index=True)
```

```python
all_data = Feb_data
all_data
```

```python
all_data.dtypes
```

```python
all_data['started_at'] = pd.to_datetime(all_data['started_at'])
```

```python
all_data['ended_at'] = pd.to_datetime(all_data['ended_at'])
```

```python
all_data.dtypes
```

calculating trip duration

```python
all_data['trip duration'] =all_data['ended_at'] - all_data['started_at']
```

```python
all_data.head()
```

```python

all_data['trip duration mins'] = (all_data['trip duration'].dt.total_seconds())/60
```

```python
all_data.head()
```

```python
all_data['startmonth'] = all_data['started_at'].dt.month
all_data['endmonth'] = all_data['ended_at'].dt.month
```

```python
all_data['endmonth'].value_counts()
```

```python
output_data_file = "./CSV/citibike_combined_df.csv"
all_data.to_csv(output_data_file, index = False)
```
